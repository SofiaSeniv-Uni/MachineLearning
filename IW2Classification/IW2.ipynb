```python
data_dir = pathlib.Path("images")
image_paths = list(data_dir.glob("*.jpg"))
```

```python
# Функція для зчитування та попередньої обробки зображень
def load_and_preprocess_grayscale_image(image_path):
    # Зчитування зображення з файлу
    image = tf.io.read_file(image_path)
    # Декодування зображення в чорно-білий тензор
    image = tf.image.decode_jpeg(image, channels=1)  # Один канал для чорно-білого зображення
    # Зміна розміру зображення (наприклад, 224x224)
    image = tf.image.resize(image, [224, 224])
    # Нормалізація значень пікселів до діапазону [0, 1]
    image /= 255.0
    return image

# Створення потоку даних (Dataset) з усіх шляхів до зображень
dataset = tf.data.Dataset.from_tensor_slices([str(p) for p in image_paths])
# Застосування попередньої обробки до кожного зображення
dataset = dataset.map(load_and_preprocess_grayscale_image)

```
